```python
import pandas as pd
import numpy as np
import pylab as pl
from urllib.request import urlopen
import matplotlib.pylab as plt
import io
import zipfile
try:
    import urllib2 as urllib
except ImportError:
    import urllib.request as urllib
```

```python
url ='http://databank.worldbank.org/data/download/INDODAPOER_csv.zip'
```

```python
z = zipfile.ZipFile( 
        io.BytesIO( 
            urllib.urlopen(url).read() # Loads from url
        ) # Converts to bytes (from unicode) 
    ) # load zip file
print(z.namelist())
```

```python
filename = z.namelist()[0]
df = pd.read_csv(z.open(filename))
```
